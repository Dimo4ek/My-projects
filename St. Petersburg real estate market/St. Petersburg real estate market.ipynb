```python
from io import BytesIO

import matplotlib.pyplot as plt
import pandas as pd
import requests
```

Напишем функцию, чтобы прочитать и сохранить данные из Google Sheets

```python
def table(spreadsheet_id):
    spreadsheet_id = spreadsheet_id
    file_name = 'https://docs.google.com/spreadsheets/d/{}/export?format=csv'.format(spreadsheet_id)
    r = requests.get(file_name)
    table = pd.read_csv(BytesIO(r.content))
    return table
```

Заменим пропуску в столбцах `is_apartment` и `balcony` на более подходящие логические значения. 
<br>Для балконов пропуски заменим на 0, тк очевидно, что если продавец не указывает информацию о балконе, то его вероятнее всего нет.
<br>Все пропуски в обозначении апартаментов заменим на `False`, тк если продавец не указывает этот параметр, то  вероятнее всего недвижимость не относится к апартаментам.

```python
df['balcony'] = df['balcony'].fillna(0)  # Заменяем пропуски в балконах на значение 0
df['is_apartment'] = df['is_apartment'].fillna(False) # Заменяем пропуски обозначениях апартаментов на False
df['locality_name'] = df['locality_name'].fillna('Другое')  
```

Заменим пропуски в столбце `floors_total` из столбца этаж квартиры:

```python
df['floors_total'] = df['floors_total'].fillna(df['floor'])# Заменяем пропуски этажности здания значение из столбца этаж квартиры
# Заменять пропуски в длительности размещения не будем, тк дата слишком специфична и в целом столбец в исследовании не используем
```

Заменим пропуски в столбце `kitchen_area` для апартаментов нулевыми значениями, для остальных объявлений разницей общей площади и жилой площади, а если останутся пропуски, то уже их заменим медианой:

```python
# апартаментов с пропусками в площади кухни 15 строк
df.query('is_apartment == True and kitchen_area == "NaN"')['kitchen_area'] 
```

```python
def kitchen_area(row): # создадим функцию которой передаём сортировку значений
    if row['is_apartment'] == True:
        return 0
    else:
        return row['kitchen_area']
df['kitchen_area'] = df.apply(kitchen_area, axis = 1) # заменим значения в столбце площадь кухни через apply согласно логике функции

df['kitchen_area'] = df['kitchen_area'].fillna(df['total_area'] - df['living_area']) # Площадь кухни равна разности общей площади помещения и жилой площади

df['kitchen_area'] = df['kitchen_area'].transform(lambda x: x.fillna(x.median()))# Заменяем пропуски площади кухни на медианное значение
```

Заменим пропуски в столбце `living_area` разницей общей площади и площади кухни:

```python
df['living_area'] = df['living_area'].fillna(df['total_area'] - df['kitchen_area'])
```

Заменим пропуски в столбце `ceiling_height` на медианное значение, предварительно заменив слишком  большие (более 5 метров) и слишком маленькие значения (менее 2,5 метров, тк по данным открытых источников - это минимальная высота потолков в пятиэтажках):

```python
df['ceiling_height'].describe() #посмотрим параметры данных столбца высота потолков
```

```python
def ceiling_height(row): # создадим функцию которой передаём сортировку значений
    if row['ceiling_height'] < 2.50:
        return 2.50
    elif row['ceiling_height'] > 5.00:
        return 5.00
    else:
        return row['ceiling_height']
    
df['ceiling_height'] = df.apply(ceiling_height, axis = 1) # заменим значения в столбце высота потолков через apply согласно логике функции
```

```python
df['ceiling_height'].describe()
```

```python
df['ceiling_height'] = df['ceiling_height'].transform(lambda x: x.fillna(x.median())) # Заменяем пропуски в высоте потолка на медианное значение
```

### Приведём данные к нужным типам

Заменим вещественные данные на целочисленные:

```python
df['last_price'] = df['last_price'].astype('int') # заменим вещественные данные на числовые используя метод astype в цене объекта 
df['floors_total'] = df['floors_total'].astype('int') # заменим вещественные данные на числовые в этажности здания
df['balcony'] = df['balcony'].astype('int') # заменим вещественные данные на числовые в числе балконов
```

Преобразуем формат стобца даты публикации объявления в `datetime`:

```python
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S') # меняем формат в стобце даты публикации объявления на datetime 
df['first_day_exposition'] = df['first_day_exposition'].dt.round('1D') # Округляем значения в столбце дата публикации объявления до суток
```

```python
df.info() # проверим форматы столбцов после преобразований
```

## Расчёты и добавление результатов в таблицу

Расчитаем цену квадратного метра для каждой квартиры и добавим столбец в таблицу:

```python
df['price_meter'] = df['last_price'] / df['total_area'] # считаем цену квадратного метра для каждого объявления и вносим в новый столбец
df['price_meter'] = df['price_meter'].astype('int') # заменим вещественные данные на числовые в цене за квадратный метр
```

Добавим в таблицу для даты публикации каждого объявления день недели, месяц и год:

```python
df['weekday'] = df['first_day_exposition'].dt.weekday # добавляем колонку номера дня недели
df['month'] = df['first_day_exposition'].dt.month # добавляем колонку номера месяца
df['year'] = df['first_day_exposition'].dt.year # добавляем колонку года
```

Добавим в таблицу вариант этажа - первый, последний, другой:

```python
def floor_r(row): # создадим функцию которой передаём сортировку значений (первый, последний, другой) для нового столбца floor_category
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
    
df['floor_category'] = df.apply(floor_r, axis = 1) # с помощью функции apply создадим столбец floor_category используя функцию в которую прописали условия для сортировки
```

Посчитаем и добавим в таблицу отношение жилой и общей площади:

```python
df['living_total_area'] = df['living_area'] / df['total_area'] # считаем отношение жилой площади объекта к общей площади
```

Посчитаем и добавим в таблиц отношение площади кухни к общей:

```python
df['kitchen_total_area'] = df['kitchen_area'] / df['total_area'] # считаем отношение площади кухни объекта к общей площади
```

**Вывод**

```python
def floor_r_2(row): # создадим функцию которой передаём сортировку значений (первый = 1, последний = 3, другой = 2) для нового столбца floor_category_2
    if row['floor'] == 1:
        return 1
    elif row['floor'] == row['floors_total']:
        return 3
    else:
        return 2
    
df['floor_category_2'] = df.apply(floor_r_2, axis = 1) # с помощью функции apply создадим столбец floor_category_2 используя функцию в которую прописали условия для сортировки

df.plot(x='floor_category_2', y='last_price', ylim=(0, 80000), kind='scatter', figsize=(15, 15), grid=True)
```
